```python
!pip install -q gradio

import gradio as gr

# 1. 데이터베이스
ORIGINAL_DB = [
    {
        "word": "flourish",
        "meaning": "번창하다, 잘 자라다",
        "sentence": "The plants will [ f_______ ] in full sun.",
        "definition": "To grow or develop successfully; to thrive.",
        "hint": "f"
    },
    {
        "word": "reveal",
        "meaning": "드러내다, 밝히다",
        "sentence": "The test results will [ r_______ ] the truth.",
        "definition": "To make known a secret or something that was hidden.",
        "hint": "r"
    },
    {
        "word": "encounter",
        "meaning": "마주치다, 맞닥뜨리다",
        "sentence": "I didn't expect to [ e_______ ] a bear in the woods.",
        "definition": "To meet someone or experience something unexpectedly.",
        "hint": "e"
    },
    {
        "word": "maintain",
        "meaning": "유지하다, 관리하다",
        "sentence": "You should [ m_______ ] a healthy weight.",
        "definition": "To continue to have something; to keep it in good condition.",
        "hint": "m"
    },
    {
        "word": "transform",
        "meaning": "변형시키다, 완전히 바꾸다",
        "sentence": "Technology can [ t_______ ] the way we live.",
        "definition": "To change the form, appearance, or character of something completely.",
        "hint": "t"
    }
]

current_quiz_list = list(ORIGINAL_DB)
incorrect_list = []
pointer = 0
round_count = 1

# 2. 로직 제어 함수들 (리턴 순서도 UI에 맞춰 Definition -> Sentence로 변경)
def load_quiz():
    global current_quiz_list, pointer, round_count

    if not current_quiz_list:
        return (
            "종료되었습니다.",
            "🎉 모든 문제를 마스터했습니다! 완벽한 Vocabulary Resonance 달성!",
            "학습 완료 👍",
            "", "", "",
            gr.update(visible=False),
            gr.update(visible=False),
            gr.update(visible=False)
        )

    quiz = current_quiz_list[pointer]
    status = f"🔄 {round_count}회독 학습 중 | 남은 문제 수: {pointer + 1} / {len(current_quiz_list)}"

    # 순서 변경: quiz["definition"]이 먼저 가고, 그 다음 quiz["sentence"]가 갑니다.
    return (
        quiz["definition"],
        quiz["sentence"],
        status,
        "", "", "",
        gr.update(visible=True),
        gr.update(visible=True),
        gr.update(visible=False)
    )

def check_answer(user_input):
    global current_quiz_list, incorrect_list, pointer

    if not current_quiz_list:
        return "", "", gr.update(), gr.update(), gr.update()

    quiz = current_quiz_list[pointer]
    correct_word = quiz["word"]

    if user_input.strip().lower() == correct_word.lower():
        feedback = f"🎉 정답입니다!: {correct_word} ({quiz['meaning']})"
        resonance = f"💡 [Vocabulary Resonance]: 형태(Form)-의미(Meaning) 연결 성공!"
        return feedback, resonance, gr.update(visible=False), gr.update(visible=False), gr.update(visible=True)
    else:
        feedback = f"❌ 아쉽습니다. 다시 생각해보세요!"
        resonance = f"💡 [힌트]: '{quiz['hint']}'로 시작하는 {len(correct_word)}글자 단어입니다."

        if quiz not in incorrect_list:
            incorrect_list.append(quiz)

        return feedback, resonance, gr.update(visible=True), gr.update(visible=True), gr.update(visible=False)

def skip_and_show_answer():
    global current_quiz_list, incorrect_list, pointer

    if not current_quiz_list:
        return "", "", gr.update(), gr.update(), gr.update()

    quiz = current_quiz_list[pointer]
    correct_word = quiz["word"]

    feedback = f"⏭️ 건너뛰기 완료 | 정답은 [ {correct_word} ] ({quiz['meaning']}) 입니다."
    resonance = f"💡 [Input 제공]: 정답을 인지하며 새로운 Cue strength를 형성하는 과정입니다."

    if quiz not in incorrect_list:
        incorrect_list.append(quiz)

    return feedback, resonance, gr.update(visible=False), gr.update(visible=False), gr.update(visible=True)

def move_next():
    global current_quiz_list, incorrect_list, pointer, round_count

    pointer += 1

    if pointer >= len(current_quiz_list):
        if incorrect_list:
            current_quiz_list = list(incorrect_list)
            incorrect_list = []
            pointer = 0
            round_count += 1
            gr.Info(f"📢 {round_count}회독 시작! 틀린 문제 복습 세트로 진입합니다.")
        else:
            current_quiz_list = []

    return load_quiz()

# 3. Gradio UI 레이아웃
with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown("# 🧠 UCM 기반 오답 순환형 레조넌스 빌더 (단서 순서 최적화)")
    gr.Markdown("### 📝 영영 정의(개념 단서)를 먼저 파악한 뒤, 빈칸 문맥(부분 자막 단서)을 조합하여 정답을 유추하세요.")

    with gr.Row():
        # 왼쪽 창: 언어적 단서 제공 (Linguistic Cues) - 순서 조정 완료
        with gr.Column(scale=1):
            gr.Markdown("### 🔍 제공된 언어적 단서 (Linguistic Cues)")
            q_definition = gr.Textbox(label="1. 개념 단서 (English Definition)", interactive=False, lines=2)
            q_sentence = gr.Textbox(label="2. 부분 자막 단서 (Partial Captioning)", interactive=False, lines=2)
            status_msg = gr.Markdown("로딩 중...")

        # 오른쪽 창: 학생 인출 및 상호작용
        with gr.Column(scale=1):
            gr.Markdown("### ✍️ 인출 및 활성화 (Active Retrieval)")
            user_input = gr.Textbox(label="나의 답변 입력", placeholder="빈칸에 들어갈 영어 단어를 타이핑하세요.")

            with gr.Row():
                submit_btn = gr.Button("정답 확인 ✔️", variant="primary")
                skip_btn = gr.Button("모르겠어요 ⏭️", variant="stop")
                next_btn = gr.Button("다음 문제로 ➡️", variant="secondary", visible=False)

    # 하단 창: SLA 인지 효과 피드백
    with gr.Row():
        feedback_box = gr.Textbox(label="채점 결과", interactive=False)
        resonance_box = gr.Textbox(label="SLA 인지 과학적 분석 (Cognitive Analysis)", interactive=False, lines=3)

    # --- 이벤트 흐름 연결 (출력 컴포넌트 순서도 q_definition -> q_sentence로 매핑) ---
    submit_btn.click(fn=check_answer, inputs=[user_input], outputs=[feedback_box, resonance_box, submit_btn, skip_btn, next_btn])
    user_input.submit(fn=check_answer, inputs=[user_input], outputs=[feedback_box, resonance_box, submit_btn, skip_btn, next_btn])

    skip_btn.click(fn=skip_and_show_answer, inputs=[], outputs=[feedback_box, resonance_box, submit_btn, skip_btn, next_btn])

    next_btn.click(fn=move_next, inputs=[], outputs=[q_definition, q_sentence, status_msg, user_input, feedback_box, resonance_box, submit_btn, skip_btn, next_btn])

    demo.load(fn=load_quiz, inputs=[], outputs=[q_definition, q_sentence, status_msg, user_input, feedback_box, resonance_box, submit_btn, skip_btn, next_btn])

# 실행 및 공용 링크 생성
demo.launch(share=True)
```
